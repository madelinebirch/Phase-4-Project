```python
# Using Pandas pivot table to create the user-item interaction matrix
user_item_matrix = ratings_df.pivot_table(index='userId', columns='movieId', values='rating')
user_item_matrix
```

```python
# Fill missing values with zeros
user_item_matrix = user_item_matrix.fillna(0)
```

```python
user_item_matrix
```

```python
import numpy as np
from scipy.sparse import coo_matrix
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
```

```python
# Convert user-item matrix to COO format for Surprise
coo_user_item_matrix = coo_matrix(user_item_matrix)
```

```python
#Define a reader and create a Surprise dataset from the COO user-item matrix.
reader = Reader(line_format='user item rating', sep='\t', rating_scale=(0, 5))
data = Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], reader)
```

```python
#Split the data into training and testing sets for model evaluation.
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
#Create an SVD model and train it using the training set.
model = SVD()
model.fit(trainset)
```

accuracy:

```python
from surprise import accuracy

predictions = model.test(testset)
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
```

```python
def get_top_n_recommendations(user_id, n=5):
    user_items = user_item_matrix.loc[user_id]
    unrated_items = user_items[user_items == 0].index
    recommendations = []

    for item_id in unrated_items:
        predicted_rating = model.predict(user_id, item_id).est
        recommendations.append((item_id, predicted_rating))

    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations[:n]
```

testing on a user:

```python
user_id = 1  # Replace with the user's ID you want to test
recommendations = get_top_n_recommendations(user_id, n=5)

for rank, (item_id, predicted_rating) in enumerate(recommendations, start=1):
    print(f"Rank {rank}: Movie ID {item_id} (Predicted Rating: {predicted_rating:.2f})")
```
